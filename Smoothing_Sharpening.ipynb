```
import numpy as np 
import cv2
import matplotlib.pyplot as plt  

def Filter_Image(Image,Filter):
  # Shape of filter and image
  a , b =  Filter.shape
  i_x , i_y = Image.shape

  # Output shape 
  x = i_x - a + 1
  y = i_y - b + 1

  # New Image
  F_N = np.ones((x,y))
  for i in range(x):
    for j in range(y):
      F_N[i][j] = np.sum(Image[i:i+a,j:j+b]*Filter)
  Final = np.pad(F_N, mode='constant', pad_width=1)    
  return Final

K1 = np.array([[0,0,-1],[0,0,0],[1,0,0]])
K2 = np.array([[0,0.1,0],[0.2,0.4,0.2],[0,0.1,0]])
K3 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

I1 = cv2.imread("/content/I1.JPG", cv2.IMREAD_GRAYSCALE)
I2 = cv2.imread("/content/I2.JPG", cv2.IMREAD_GRAYSCALE)

print('1st Image')
plt.figure(figsize=(3,3))
plt.imshow(I1, cmap='gray')
plt.show()

print('Kernal : K1')
o1 = Filter_Image(I1,K1)
plt.figure(figsize=(3,3))
plt.imshow(o1, cmap='gray')
plt.show()

print('Kernal : K2')
o2 = Filter_Image(I1,K2)
plt.figure(figsize=(3,3))
plt.imshow(o2, cmap='gray')
plt.show()

print('Kernal : K3')
o3 = Filter_Image(I1,K3)
plt.figure(figsize=(3,3))
plt.imshow(o3, cmap='gray')
plt.show()

print('2nd Image')
plt.figure(figsize=(3,3))
plt.imshow(I2, cmap='gray')
plt.show()

print('Kernal : K1')
o1 = Filter_Image(I2,K1)
plt.figure(figsize=(3,3))
plt.imshow(o1, cmap='gray')
plt.show()

print('Kernal : K2')
o2 = Filter_Image(I2,K2)
plt.figure(figsize=(3,3))
plt.imshow(o2, cmap='gray')
plt.show()

print('Kernal : K3')
o3 = Filter_Image(I2,K3)
plt.figure(figsize=(3,3))
plt.imshow(o3, cmap='gray')
plt.show()


```

```
# Average and weighted average filters
Filter1 = (1/49)*np.ones((7,7))
D_Conv(im2,Filter1)

Filter2 = (1/25)*np.ones((5,5))
Filter2 = Filter2/np.sum(Filter2)
D_Conv(im2,Filter2)

Filter3 = np.array([[1,2,1],[2,4,2],[1,2,1]])
Filter3 = Filter3/np.sum(Filter3)
D_Conv(im2,Filter3)

Filter4 = np.array([[1,2,4,2,1],[2,4,6,4,2],[4,6,8,6,4],[1,2,4,2,1],[2,4,6,4,2]])
Filter4 = Filter4/np.sum(Filter4)
D_Conv(im2,Filter4)

# Gaussian Filter
T1 = np.array([1,2,1])
T2 = T1.reshape(3,1)
Filter5 = T2*T1
D_Conv(im2,Filter5)

T1 = np.array([1,4,6,4,1])
T2 = T1.reshape(5,1)
Filter6 = T2*T1
D_Conv(im2,Filter6)
```

```
# Non Linear Smoothening Filter
def Min_Filter(Image,Filter):
  a , b =  Filter
  x , y = Image.shape
  o_x = x - a + 1
  o_y = y - b + 1
  new_I = np.zeros((o_x,o_y))
  for i in range(o_x):
    for j in range(o_y):
      t = Image[i:i+a,j:j+b]
      x = np.min(t)
      new_I[i][j] = x
  plt.imshow(new_I, cmap='gray')
  plt.show()

#Min_Filter(im2,(3,3))

def Max_Filter(Image,Filter):
  a , b =  Filter
  x , y = Image.shape
  o_x = x - a + 1
  o_y = y - b + 1
  new_I = np.zeros((o_x,o_y))
  for i in range(o_x):
    for j in range(o_y):
      t = Image[i:i+a,j:j+b]
      x = np.max(t)
      new_I[i][j] = x
  plt.imshow(new_I, cmap='gray')
  plt.show()

#Max_Filter(im2,(3,3))

def Median_Filter(Image,Filter):
  a , b =  Filter
  x , y = Image.shape
  o_x = x - a + 1
  o_y = y - b + 1
  new_I = np.zeros((o_x,o_y))
  for i in range(o_x):
    for j in range(o_y):
      t = Image[i:i+a,j:j+b]
      x = np.median(t)
      new_I[i][j] = x
  plt.figure(figsize=(3,3))    
  plt.imshow(new_I, cmap='gray')
  plt.show()

Median_Filter(im2,(3,3))
Median_Filter(im2,(5,5))
```

```
# Edge Detectors and Sharpening Filters(2D)

# Laplacian Filters(2D)
Filter1 = np.array([[0,1,0],[1,-4,1],[0,1,0]])
D_Conv(im2,Filter1)

Filter2 = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
D_Conv(im2,Filter2)

Filter3 = np.array([[1,1,1],[1,-8,1],[1,1,1]])
D_Conv(im2,Filter3)

Filter4 = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
D_Conv(im2,Filter4)

# Composite Laplacian Filters(2D)
Filter5 = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
D_Conv(im2,Filter5)

Filter6 = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
D_Conv(im2,Filter6)

# High Boost Filter(2D)
Filter7 = np.array([[-1,-1,-1],[-1,10,-1],[-1,-1,-1]])
D_Conv(im2,Filter7) 

Filter8 = np.array([[0,-1,0],[-1,6,-1],[0,-1,0]])
D_Conv(im2,Filter8) 

# Laplacian of Gaussian (LoG)(2D)
Filter9 = np.array([[0,0,1,0,0],[0,1,2,1,0],[1,2,-16,2,1],[0,0,1,0,0],[0,1,2,1,0]])
D_Conv(im2,Filter9) 

```

```
# Edge Detectors and Sharpening Filters(1D)

# Sobel Operator(1D)
Filter1 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
D_Conv(im2,Filter1)

Filter2 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
D_Conv(im2,Filter2)

# Prewitts Operator(1D)
Filter3 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
D_Conv(im2,Filter3)

Filter4 = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
D_Conv(im2,Filter4)

# Canny Edge Detector
Filter5 = cv2.Canny(im2,100,150)
plt.imshow(Filter5, cmap='gray')
plt.show()

# Kayyali Edge Filter
Filter6 = np.array([[6,0,-6],[0,0,0],[-6,0,6]])
D_Conv(im2,Filter6)

Filter7 = np.array([[-6,0,6],[0,0,0],[6,0,-6]])
D_Conv(im2,Filter7)
```

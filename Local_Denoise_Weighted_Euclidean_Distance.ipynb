```
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
import warnings
import pickle
warnings.filterwarnings("ignore")
```

```
# Image plot
def img_plot(im):
  plt.figure(figsize=(5,5))
  plt.imshow(im,cmap='gray')
  plt.show()

# PSNR
def PSNR(I1,I2): 
    MSE = I1 - I2
    MSE = MSE**2
    MSE = np.mean(MSE) 
    p = 20*math.log10(255.0) - 10*math.log10(MSE+0.00001)
    return p 

# Local Denoise
def NonLocalDenoise(im,shape = (3,3),h=2):
  (a,b) =  shape
  imb = cv2.copyMakeBorder(im, a//2, a//2, b//2, b//2,cv2.BORDER_REPLICATE)
  imnew = im
  ha = a//2
  hb = b//2
  Nw = (2 * a + 1) 

  for i in range(a,im.shape[0]-a,a):
    for j in range(b,im.shape[1]-b,b):

      t =  imb[i-ha:i+ha+1,j-hb:j+hb+1]
      normal = 0
      weight = 0
      tf = t.flatten() 
      Nw = (2 * a + 1) ** 2
      # for k in range(0,im.shape[0]-a,1):
      #   for l in range(0,im.shape[1]-b,1):
      for k in range(i-20,i+20,1):
        for l in range(j-20,j+20,1):
        
          if k == i and l ==j:
              continue
          else:  
            if k<a:
              k =a
            if l<b:
              l =b
            if k>imb.shape[0]-a:
              k = imb.shape[0]-a
            if l>imb.shape[1]-b:
              l = imb.shape[1]-b  

          p = imb[k-hb:k+ha+1,k-hb:k+hb+1]
          
          pf = p.flatten()
            
          temp = np.sum((t-p)**2)
          temp = temp - (t[ha,hb] - p[ha,hb])
          temp = math.exp(-temp/(h**2))  
          weight = weight  + imb[k,l]*temp
          normal = normal + temp
                
      imnew [i-ha,j-hb] = weight/normal

  return imnew
```

```
path = '/content/drive/My Drive/Projects/DIP_Assignments/Test_Images/5.1.12.tiff'
im = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
img_plot(im)
print(im.shape)
```

```
# Local Denoising Filtering  : shape=(5,5),var=10 
psnr = []
img = []
for i in range(1):
  noise = np.random.normal(0,10**0.5,size= im.shape)
  imn = im + noise
  imnew  =  NonLocalDenoise(imn,shape = (9,59))
  img_plot(imnew)
  img_plot(im)
```
